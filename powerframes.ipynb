```python
# add gender to the dataframe

#gender dictionary

gender_dict = {}

male_characters = ['Moroso','Sophocles', 'Tranio', 'Jaques', 'Doctor', 'Pedro', 'Petruchio', 'Rowland', 'Servant', 'Sophocles', 'Watchmen','Petronius'] 
female_characters = ['Bianca', "CityWife", "CountryWife", "Livia", "Maids", "Maria"] 

for female_character in female_characters:
  gender_dict[female_character] = "F"

for male_character in male_characters:
  gender_dict[male_character] = "M"

# repeat for each character

#somehow add this to the dataframe..

print(list(gender_dict.items())[:1]) # print first item in dictionary.

#convert dictionary to dataframe

df_gender = pd.DataFrame([gender_dict])

#transpose dataframe
df_gender_transposed = df_gender.T # or df1.transpose()

df_gender_transposed.columns =['gender']

df_gender_transposed
```

```python
#map gender dictionary onto dataframe


#for speakers
df_tamertamed['speaker_gender']= df_tamertamed['speaker'].map(gender_dict)

#also add for characters in character list?
```

```python
# Extracting key-value of dictionary in variables
df_tamertamed['sd_gender'] = " "

for index, row in df_tamertamed.iterrows():
    sd_gender_list = []
    character_list = row['character_list']
    for word in character_list:
        if gender_dict[word][0] == 'M':
          sd_gender_list.append('M')
        elif gender_dict[word][0] == 'F':
          sd_gender_list.append('F')
        else:
          sd_gender_list.append('N')
    df_tamertamed.at[index, 'sd_gender'] = sd_gender_list
  #Have to deal with the list somehow...
```
