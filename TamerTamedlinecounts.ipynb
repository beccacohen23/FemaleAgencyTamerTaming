```python
import pandas as pd
import json

#convert JSON back to dataframe for analysis:

df_tamertamed_final = pd.read_json('powerframes_final_df_tamertamed_withgendercount.json')

df_tamertamed_final
```

```python
for index, row in df_tamertamed_final.iterrows():
  word_count = 0
  line_count = 1
  sentence_count = 0
  sentences = row['text'].split('.')
  for sentence in sentences:
    sentence_count +=1
  df_tamertamed_final.at[index, 'line_count'] = line_count
  df_tamertamed_final.at[index, 'sentence_count'] = sentence_count

for index, row in df_tamertamed_final.iterrows():
  word_count = 0
  words = row['text'].split(' ')
  for word in words:
    word_count +=1
  df_tamertamed_final.at[index, 'word_count'] = word_count

for index, row in df_tamertamed_final.iterrows():
  verb_count = 0
  verb_lemmas = row['verb_lemmas']
  for verb_lemma in verb_lemmas:
    verb_count +=1
    df_tamertamed_final.at[index, 'verb_count'] = verb_count 


```

```python

###########
#BY GENDER#
###########
female_filter = df_tamertamed_final['speaker_gender'] == 'F'
male_filter = df_tamertamed_final['speaker_gender'] == 'M'

tamertamed_women = df_tamertamed_final[female_filter]
tamertamed_men = df_tamertamed_final[male_filter]

#line counts
tamertamed_men['line_count'].sum()
tamertamed_women['line_count'].sum()
#word counts
tamertamed_men['word_count'].sum()
tamertamed_women['word_count'].sum()
#verb counts
tamertamed_men['verb_count'].sum()
tamertamed_women['verb_count'].sum()
#sentence counts
tamertamed_men['sentence_count'].sum()
tamertamed_women['sentence_count'].sum()


d = {'line_count': [tamertamed_men['line_count'].sum(),tamertamed_women['line_count'].sum() ], 'word_count': [tamertamed_men['word_count'].sum(), tamertamed_women['word_count'].sum()], 'verb_count': [tamertamed_men['verb_count'].sum(), tamertamed_women['verb_count'].sum()], 'sentence_count': [tamertamed_men['sentence_count'].sum(), tamertamed_women['sentence_count'].sum()]}

df = pd.DataFrame(data=d)

df.index = ['tamertamed_men', 'tamertamed_women']

df
```
