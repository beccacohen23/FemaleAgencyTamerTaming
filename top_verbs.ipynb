```python
#create a new dataframe with just speaker and lemma verbs
df2 = df_tamertamed_final[['speaker','verb_lemmas','speaker_gender']]

df2
```

```python
#for item in character list
  #if new verb append to list
  #count most frequent verbs
  #slice data frame, count most frequent verbs by gender, by speaker


male_characters = ['Moroso','Sophocles', 'Tranio', 'Jaques', 'Doctor', 'Pedro', 'Petruchio', 'Rowland', 'Servant', 'Sophocles', 'Watchmen','Petronius'] 
female_characters = ['Bianca', 'CityWife', "CountryWife", "Livia", "Maids", "Maria"] 



#create list of all verbs used?
verb_lemma_list = []
for index, row in df2.iterrows(): 
  for word in row['verb_lemmas']:
    if word not in verb_lemma_list:
      verb_lemma_list.append(word)

#  print(verb_lemma_list)

  for index2, row2 in df2.iterrows():
    for male_character in male_characters:
      verb_count = 0
      if male_character == row['speaker']:
        for verb in row['verb_lemmas']:
          if verb in verb_lemma_list:
            verb_count +=1
      #increase counter by one for each verb and each time that speaker uses that verb
    df2.loc[index,'verb_count'] = verb_count

df2

#try at next?



      
```
