```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/honors thesis/punctuation/df_tamertamedwithpunctuation.csv')
```

```python

from __future__ import unicode_literals
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def flatten_subtree(st):
    return ''.join([w.text_with_ws for w in list(st)]).strip()
```

```python
#subjects = []
df['sentence']= df['sentence'].astype(str)
for index, row in df.iterrows():
  doc = nlp(row['sentence'])
  subjects = []
  for word in doc:
    #subjects = []
    if word.dep_ in ('nsubj', 'nsubjpass'):
      #obj of that verb also?
      subjects.append(word.text)
      #df.at[index,'subjects'] = subjects[0]
      df.at[index,'subjects'] = subjects
```
